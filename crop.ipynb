```python
import numpy as np
import seaborn as sns
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import MinMaxScaler, StandardScaler, Normalizer, Binarizer, LabelEncoder
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
import tkinter as tk
from functools import partial
from typing import List

def read_in_and_split_data(data,target):
    X=data.drop(target,axis=1)
    y=data[target]
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
    return X_train,X_test,y_train,y_test
def predict_crop(composition:List[int],temp:int,humidity:int,ph:int,rainfall:int):
    a,b,c=composition
    d,e,f,u=temp,humidity,ph,rainfall
    df=pd.read_csv('Crop_recommendation.csv')
    #Q1=df.quantile(0.25)
    #Q3=df.quantile(0.75)
    #IQR=Q3-Q1
    #df_out=df[~((df<(Q1-0.5*IQR))|(df>(Q3+1.5*IQR))).any(axis=1)]
    target='label'
    X_train,X_test,y_train,y_test=read_in_and_split_data(df,target)
    #pipeline=make_pipeline(StandardScaler(),GaussianNB())
    pipeline=make_pipeline(StandardScaler(),KNeighborsClassifier())
    model=pipeline.fit(X_train,y_train)
    y_pred=model.predict([[a,b,c,d,e,f,u]])
    result_label.config(text=f"We have profit, if we cultivate this crop {y_pred}")

def on_submit():
    composition = [eval(n.get()) for n in composition_entries]
    temp = eval(temp_entry.get())
    humidity = eval(humidity_entry.get())
    seed_weight = eval(seed_entry.get())
    rainfall = eval(rain_entry.get())
    predict_crop(composition, temp, humidity, seed_weight, rainfall)

    
# create GUI
root = tk.Tk()
root.title("CROP RECOMMENDATION SYSTEM")
# composition
tk.Label(root, text="COMPOSITION OF SOIL").grid(row=10, column=0, sticky="w")
composition_labels = ["N-Nitrogen", "P-Phosphorous", "K-Potassium"]
composition_entries = []
for i, label in enumerate(composition_labels):
    tk.Label(root, text=label).grid(row=i+1, column=0, sticky="w")
    n = tk.Entry(root)
    n.grid(row=i+1, column=1)
    composition_entries.append(n)
# temperature
tk.Label(root, text="Temperature of the region").grid(row=4, column=0, sticky="w")
temp_entry = tk.Entry(root)
temp_entry.grid(row=4, column=1)
# humidity
tk.Label(root, text="Humidity in H mm").grid(row=5, column=0, sticky="w")
humidity_entry = tk.Entry(root)
humidity_entry.grid(row=5, column=1)
# seed weight
tk.Label(root, text="pH Value").grid(row=6, column=0, sticky="w")
seed_entry = tk.Entry(root)
seed_entry.grid(row=6, column=1)
# rainfall
tk.Label(root, text="RainFall in mm").grid(row=7, column=0, sticky="w")
rain_entry = tk.Entry(root)
rain_entry.grid(row=7, column=1)
# submit button
submit_button = tk.Button(root, text="Submit", command=on_submit)
submit_button.grid(row=8, column=0, columnspan=2, pady=10)
# result
result_label = tk.Label(root, text="")
result_label.grid(row=9, column=0, columnspan=2)

root.mainloop()
```
